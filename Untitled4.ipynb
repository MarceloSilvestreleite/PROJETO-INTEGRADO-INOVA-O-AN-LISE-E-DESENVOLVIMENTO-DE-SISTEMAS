```python
# Importa√ß√£o das bibliotecas necess√°rias
import sqlite3  # Para trabalhar com banco de dados SQLite
import pandas as pd  # Para manipula√ß√£o e an√°lise de dados
import matplotlib.pyplot as plt  # Para cria√ß√£o de gr√°ficos
import seaborn as sns  # Para visualiza√ß√µes estat√≠sticas mais avan√ßadas

# Configurar o estilo dos gr√°ficos
plt.style.use('default')
sns.set_palette("husl")

# Passo 1: Conectar ao banco de dados SQLite
conexao = sqlite3.connect('dados_vendas.db')
cursor = conexao.cursor()

print("‚úÖ Conex√£o com banco de dados estabelecida com sucesso!")

# Criar tabela de vendas se n√£o existir
cursor.execute("""
CREATE TABLE IF NOT EXISTS vendas1 (
    id_venda INTEGER PRIMARY KEY AUTOINCREMENT,
    data_venda DATE,
    produto TEXT,
    categoria TEXT,
    valor_venda REAL
)
""")

print("‚úÖ Tabela 'vendas1' criada ou j√° existente!")

# Dados de vendas do √∫ltimo ano
dados_vendas = [
    ('2023-01-01', 'Produto A', 'Eletr√¥nicos', 1500.00),
    ('2023-01-05', 'Produto B', 'Roupas', 350.00),
    ('2023-02-10', 'Produto C', 'Eletr√¥nicos', 1200.00),
    ('2023-03-15', 'Produto D', 'Livros', 200.00),
    ('2023-03-20', 'Produto E', 'Eletr√¥nicos', 800.00),
    ('2023-04-02', 'Produto F', 'Roupas', 400.00),
    ('2023-05-05', 'Produto G', 'Livros', 150.00),
    ('2023-06-10', 'Produto H', 'Eletr√¥nicos', 1000.00),
    ('2023-07-20', 'Produto I', 'Roupas', 600.00),
    ('2023-08-25', 'Produto J', 'Eletr√¥nicos', 700.00),
    ('2023-09-30', 'Produto K', 'Livros', 300.00),
    ('2023-10-05', 'Produto L', 'Roupas', 450.00),
    ('2023-11-15', 'Produto M', 'Eletr√¥nicos', 900.00),
    ('2023-12-20', 'Produto N', 'Livros', 250.00)
]

# Inserir dados na tabela
cursor.executemany("""
INSERT INTO vendas1 (data_venda, produto, categoria, valor_venda)
VALUES (?, ?, ?, ?)
""", dados_vendas)

# Confirmar as inser√ß√µes
conexao.commit()

print("‚úÖ Dados inseridos com sucesso!")
print(f"‚úÖ Total de {len(dados_vendas)} registros inseridos na tabela!")

# Carregar dados do banco para um DataFrame do Pandas
df_vendas = pd.read_sql_query("SELECT * FROM vendas1", conexao)

# Converter a coluna de data para formato datetime
df_vendas['data_venda'] = pd.to_datetime(df_vendas['data_venda'])

# Exibir informa√ß√µes b√°sicas do DataFrame
print("üìä VISUALIZA√á√ÉO INICIAL DOS DADOS:")
print("=" * 50)
print(df_vendas.head())  # Primeiras 5 linhas

print("\nüìã INFORMA√á√ïES DO DATAFRAME:")
print("=" * 50)
print(df_vendas.info())

print("\nüìà ESTAT√çSTICAS DESCRITIVAS:")
print("=" * 50)
print(df_vendas.describe())

# An√°lise 1: Vendas por categoria
vendas_por_categoria = df_vendas.groupby('categoria')['valor_venda'].sum()

# An√°lise 2: Vendas mensais
vendas_por_mes = df_vendas.resample('M', on='data_venda')['valor_venda'].sum()

# An√°lise 3: Produto mais vendido por valor
produto_maior_venda = df_vendas.loc[df_vendas['valor_venda'].idxmax()]

print("üìä AN√ÅLISE DE VENDAS POR CATEGORIA:")
print("=" * 40)
print(vendas_por_categoria)

print("\nüìÖ VENDAS MENSais (2023):")
print("=" * 40)
print(vendas_por_mes)

print("\nüèÜ PRODUTO COM MAIOR VALOR DE VENDA:")
print("=" * 40)
print(f"Produto: {produto_maior_venda['produto']}")
print(f"Categoria: {produto_maior_venda['categoria']}")
print(f"Valor: R$ {produto_maior_venda['valor_venda']:.2f}")

# Gr√°fico 1: Vendas por Categoria (Barras)
plt.figure(figsize=(10, 6))
sns.barplot(x=vendas_por_categoria.index, y=vendas_por_categoria.values)
plt.title('üìä Vendas por Categoria - 2023', fontsize=16, fontweight='bold')
plt.xlabel('Categoria', fontsize=12)
plt.ylabel('Valor Total de Vendas (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adicionar valores nas barras
for i, valor in enumerate(vendas_por_categoria.values):
    plt.text(i, valor + 50, f'R$ {valor:.2f}',
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Gr√°fico 2: Vendas Mensais (Linha)
plt.figure(figsize=(12, 6))
vendas_por_mes.plot(kind='line', marker='o', linewidth=2, markersize=8)
plt.title('üìà Vendas Mensais em 2023', fontsize=16, fontweight='bold')
plt.xlabel('M√™s', fontsize=12)
plt.ylabel('Valor Total de Vendas (R$)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Formatar eixos
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("‚úÖ Visualiza√ß√µes geradas com sucesso!")

# Fechar conex√£o com o banco de dados
conexao.close()

print("‚úÖ Conex√£o com banco de dados fechada!")
print("\n" + "="*60)
print("üéØ AN√ÅLISE CONCLUSIVA:")
print("="*60)
print("""
üìå INSIGHTS OBTIDOS:

1. ‚úÖ Eletr√¥nicos √© a categoria com maior faturamento (R$ 6.100,00)
2. ‚úÖ Roupas √© a segunda categoria em performance (R$ 1.800,00)
3. ‚úÖ Livros possui o menor faturamento (R$ 900,00)
4. ‚úÖ Janeiro foi o m√™s com maior volume de vendas
5. ‚úÖ Produto A teve a venda individual de maior valor (R$ 1.500,00)

üí° SUGEST√ïES PARA A EMPRESA:

‚Ä¢ Investir em marketing para categorias de menor desempenho
‚Ä¢ Promover bundles entre eletr√¥nicos e acess√≥rios
‚Ä¢ Criar promo√ß√µes sazonais baseadas nos meses de pico
‚Ä¢ Expandir linha de produtos na categoria de Eletr√¥nicos
""")
```
